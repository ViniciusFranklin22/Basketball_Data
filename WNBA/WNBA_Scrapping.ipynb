```python
url_test = "https://stats.wnba.com/stats/leagueLeaders?LeagueID=10&PerMode=Totals&Scope=S&Season=2024&SeasonType=Regular+Season&StatCategory=PTS"
test_table = requests.get(url=url_test).json()
```

```python
api_cols = test_table["resultSet"]["headers"]
table_headers = ["SEASON","SEASON_TYPE"] + api_cols
```

```python
wnba_data = pd.DataFrame(columns=table_headers)
time_st = time.time()
number = 0

for season in seasons:
    for st in season_type:
        url = f'https://stats.wnba.com/stats/leagueLeaders?LeagueID=10&PerMode={per_mode[0]}&Scope=S&Season={season}&SeasonType={st}&StatCategory=PTS'
        main_data = requests.get(url=url,headers=headers_api).json()
        data_stats = pd.DataFrame(main_data["resultSet"]["rowSet"],columns=api_cols)
        data_seasons = pd.DataFrame({"SEASON":[season for i in range(len(data_stats))],
                                    "SEASON_TYPE":[st for x in range(len(data_stats))]})
        df_3 = pd.concat([data_stats,data_seasons],axis=1) #horizantally
        wnba_data = pd.concat([wnba_data,df_3],axis=0) #vertically
        number +=len(data_stats)
        print(f'{number}  {season} {st}')

print(f"Finished in {time.time()-time_st}")     
        
        
        
```
