```python
import pandas as pd
import requests
pd.set_option("display.max_columns",None)
import time
import numpy as np
import json
```

```python
url_test = "https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=Totals&Scope=S&Season=2016-17&SeasonType=Regular%20Season&StatCategory=PTS"
```

```python
season = ['1951-52']
i = 1952
x = 53
while season[-1] != '2023-24':
    year = f'{i}-{x}'
    season.append(year)
    i += 1
    x = 0 if x == 99 else int(x) + 1
    if x > -1 and x < 10:
        x = '0' + str(x)
```

```python
data = requests.get(url=url_test).json()
table_headers = data['resultSet']["headers"]
df_cols = table_headers
table_headers = ['YEAR','SEASON_TYPE']+ table_headers
Season_Type = ['Regular%20Season','Playoffs']
headers_api = {
    'Accept': '*/*',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept-Language': 'pt-BR,pt;q=0.9,en-US;q=0.8,en;q=0.7',
    'Connection': 'keep-alive',
    'Host': 'stats.nba.com',
    'Origin': 'https://www.nba.com',
    'Referer': 'https://www.nba.com/',
    'Sec-Ch-Ua': '"Not A(Brand";v="99", "Google Chrome";v="121", "Chromium";v="121"',
    'Sec-Ch-Ua-Mobile': '?0',
    'Sec-Ch-Ua-Platform': '"Windows"',
    'Sec-Fetch-Dest': 'empty',
    'Sec-Fetch-Mode': 'cors',
    'Sec-Fetch-Site': 'same-site',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36'
}
```

```python
nba_data = pd.DataFrame(columns=table_headers)
time_st = time.time()
number = 0
for i in season:
    for x in Season_Type:
        url = f"https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=Totals&Scope=S&Season={i}&SeasonType={x}&StatCategory=PTS"
        main_data = requests.get(url=url,headers=headers_api).json()
        data_stats =pd.DataFrame(main_data['resultSet']['rowSet'],columns=df_cols)
        data_stats_2 = pd.DataFrame({'YEAR':[i for k in range(len(data_stats))],
                            'SEASON_TYPE':[ x for y in range(len(data_stats))]})
        df_3 = pd.concat([data_stats,data_stats_2],axis=1)
        nba_data = pd.concat([nba_data,df_3],axis=0)
        number += len(data_stats)
        print(number," ",i)
print(f"Finished in {time.time()-time_st}")
        
```
